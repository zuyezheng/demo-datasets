```python
import pandas
```

```python
# see what we're dealing with
pandas.set_option('display.max_columns', None)
inventory_df = pandas.read_csv('data/sf_rent/Rent_Board_Housing_Inventory_20250314.csv')
inventory_df
```

```python
"""
The point column are values such as:
POINT (-122.394549329 37.781408195)

Split these up to get the longitude and latitude values in 2 separate columns.
"""
inventory_df[['longitude', 'latitude']] = inventory_df['point']\
    .str.extract(r'POINT \((.*)\)')[0]\
    .str.split(' ', expand=True)

# convert the new columns to numeric
inventory_df['longitude'] = pandas.to_numeric(inventory_df['longitude'])
inventory_df['latitude'] = pandas.to_numeric(inventory_df['latitude'])

inventory_df
```

```python
# clean up square footage from 501-750 Sq.Ft to 2 columns of 501 and 750
def extract_square_footage(text):
    if pandas.isna(text):
        return pandas.NA, pandas.NA
    
    # Remove common text patterns and convert to lowercase
    text = str(text).lower().replace('sq.ft', '').replace('sqft', '').replace('ft', '').replace('$', '').strip()
    
    # Handle ranges (e.g. "800-1000" or "800 - 1000")
    if '-' in text:
        parts = text.replace(' ', '').split('-')
        try:
            return float(parts[0]), float(parts[1])
        except:
            return pandas.NA, pandas.NA
    
    # Handle single values
    try:
        value = float(text.replace(',', ''))
        return value, value
    except:
        return pandas.NA, pandas.NA

# Apply the function and create new columns
inventory_df[['min_square_footage', 'max_square_footage']] = inventory_df['square_footage'].apply(
    lambda x: pandas.Series(extract_square_footage(x))
)

inventory_df
```

```python
# clean up monthly rent from $501-$750 to 2 columns with 501, 750
def extract_montly_rent(text):
    if pandas.isna(text):
        return pandas.NA, pandas.NA
    
    # Remove common text patterns and convert to lowercase
    text = str(text).lower().replace('$', '').strip()
    
    # Handle ranges (e.g. "800-1000" or "800 - 1000")
    if '-' in text:
        parts = text.replace(' ', '').split('-')
        try:
            return float(parts[0]), float(parts[1])
        except:
            return pandas.NA, pandas.NA
    
    # Handle single values
    try:
        value = float(text.replace(',', ''))
        return value, value
    except:
        return pandas.NA, pandas.NA

# Apply the function and create new columns
inventory_df[['min_monthly_rent', 'max_monthly_rent']] = inventory_df['monthly_rent'].apply(
    lambda x: pandas.Series(extract_montly_rent(x))
)

inventory_df
```

```python
# normaize the bedroom count string to a number
def extract_bedroom_count(text):
    if pandas.isna(text):
        return pandas.NA
    
    # cleanup
    text = str(text).lower().strip()

    # map numbers in words to numbers
    word_numbers = {
        'zero': 0,
        'one': 1,
        'two': 2,
        'three': 3,
        'four': 4,
        'five': 5
    }
    for word, number in word_numbers.items():
        text = text.replace(word, str(number))

    # special case for "5+"
    if text == '5+':
        return 6

    # extract only the contigous numbers e.g. "4-bedroom" -> 4, "0ah" -> 0
    number_str = ''
    for c in text:
        if c.isdigit():
            number_str += c
        else:
            break

    # cast to number
    try:
        num_bedrooms = int(number_str)
        if num_bedrooms > 5:
            return 0
        else:
            return num_bedrooms
    except:
        return 0
```

```python
# sanity check we're converting the bedroom count correctly
for bedrooms in inventory_df['bedroom_count'].unique():
    print(bedrooms, ' -> ', extract_bedroom_count(bedrooms))
```

```python
# normalize bathroom count to a number
def extract_bathroom_count(text):
    if pandas.isna(text):
        return pandas.NA
    
    # cleanup
    text = str(text).lower().strip()

    # map numbers in words to numbers
    word_numbers = {
        'one and a half': 1.5,
        'two and a half': 2.5,
        'zero': 0,
        'one': 1,
        'two': 2,
        'three': 3
    }
    for word, number in word_numbers.items():
        text = text.replace(word, str(number))

    # extract the numbers
    number_str = ''
    for c in text:
        if c.isdigit() or c == '.':
            number_str += c
        else:
            break

    # cast to number
    try:
        return float(number_str)
    except:
        return 0
```

```python
# make sure we're doing it properly
for bedrooms in inventory_df['bathroom_count'].unique():
    print(bedrooms, ' -> ', extract_bathroom_count(bedrooms))
```

```python
inventory_df.rename(columns={
    'bedroom_count': 'bedrooms_raw',
    'bathroom_count': 'bathrooms_raw'
}, inplace=True)

inventory_df['bedrooms'] = inventory_df['bedrooms_raw'].apply(extract_bedroom_count)
inventory_df['bathrooms'] = inventory_df['bathrooms_raw'].apply(extract_bathroom_count)

inventory_df
```

```python
date_columns = ['occupancy_or_vacancy_date', 'vacancy_date', 'signature_date', 'data_as_of', 'data_loaded_at', ]

# convert the string date columns to timestamps to be used in parquet
for column in date_columns:
    inventory_df[column] = pandas.to_datetime(inventory_df[column], errors='coerce')
```

```python
inventory_df.drop(columns=[
    'occupancy_or_vacancy_date_history',
    'point',
    'unique_id',
]).to_parquet(
    'data/sf_rent/rent_board_inventory.parquet',
    index=False,
    engine="pyarrow",
    coerce_timestamps="us",
    allow_truncated_timestamps=True
)
```
